```python
#HW3
#파이썬으로적분

def func(currX):
    return currX*currX #홤수를 x^2로 둠


def integral(n):
    currX = 0
    area = 0
    
    for i in range(n):
        area += func(currX) * dx #높이: func(currX), 밑변:dx 이고, 넓이를 i번만큼를 더하라.
        currX += dx #밑변을 i번 더해 i번째 직사각형의 밑변을 구한다.
    return area

start = 3
end = 6
dx = 0.0001 #3~6의 구간을 무한대로 분할 하기위한 미소값(적정값 설정)

n = round((end - start) / dx) # "dx = (b - a) /n "을 미지수 n의 식으로 나타냄
area = integral(n)

print("x^2를 3~6구간에서 적분한 값은: {}".format(area))
```

```python
#HW3
#파이썬으로미분

def func(currX):
    return currX*currX #홤수를 x^2로 둠

def differential(a, dx):
    linear = (func(a+dx)- func(a))/dx #미분공식 f(a+dx)-f(a) / dx를 사용. a점의 기울기를 구하고 싶음.
    return linear

a = 3
dx = 0.001#a+dx지점에서 a지점으로 계속 수렴하도록 미소값 설정(적정값)
res = differential(a, dx)#함수 사용하여 A점의 기울기를 구함 (a, dx가 매개체)

print("x^2를 3에서 미분한 값은: {}".format(res))
```
